# Introduction
One way to understand mathematical concepts such as neural networks is to see how they are built up from its underlying more basic mathematical building blocks. However, building up the required background in math can be a long and arduous proces. To boost our learning efficieny when can take a learning shortcut using intuitions. The trick here is to liken the mathematical concept to something the learner is already familiar with. This is similar to pitching a book to a friend as a 'heist novel where indiana jones breaks into hogwarts'. We might not know exactly what the book will be, but we have a general gist of our main character (indiana jones), the type of story (heist novel) and the setting (a magic high school). In this example we see some key properties of a good intuition:

- It connects to a concept the learner is already familiar with
- It simplifies the concept we try to learn in order to make it more easily digestible to the learner
- It provides a basis on which we can build if we want to continue learning after consuming the intuition

This article will describe a number of intuitions I find really useful. The goal is to help inexperienced people kick start their data science journey, help experienced people deepen their understanding, and help educators with usefull intuitions they can use in their classes. Where possible I try to provide the source of the intuition, any faults in representing those intuitions is my own.

# Classification can be seen as drawing optimal boundaries between training samples
Inspired by [this link](https://scikit-learn.org/stable/modules/generated/sklearn.inspection.DecisionBoundaryDisplay.html#sklearn.inspection.DecisionBoundaryDisplay)

```python
import pandas as pd
from sklearn import tree, ensemble, linear_model, naive_bayes
from plotnine import *
from support_functions import get_iris_data
theme_set(theme_bw())

def get_plot_data(classifier, tag):
    model = classifier().fit(iris_pd[['x', 'y']], iris_pd['class'])
    plot_data=pd.DataFrame({'x': grid[:,0], 'y': grid[:,1], 'class': model.predict(grid)})
    plot_data['class'] = plot_data['class'].astype('str')
    plot_data['classifier'] = tag
    return plot_data

grid, iris_pd = get_iris_data()
plot_data = pd.concat([get_plot_data(sklearn.tree.DecisionTreeClassifier, 'Decision Tree'), 
                       get_plot_data(sklearn.linear_model.LogisticRegression, 'Logistic Regression'),
                       get_plot_data(sklearn.ensemble.GradientBoostingClassifier, 'GBM'),
                       get_plot_data(sklearn.naive_bayes.GaussianNB, 'Naive Bayes')])

(
    ggplot(plot_data, aes(x='x', y='y', fill='class')) 
      + geom_tile() 
      + geom_point(iris_pd, size=3) 
      + facet_wrap('~ classifier', ncol=2) 
      + theme(figure_size=(8,6))
)
```
